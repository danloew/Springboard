Because the five target features are instantiated, the features that were used to create the target features are no longer needed. The next cell drops these features. Although violation sales started as a target variable as it was a combination of PD_CD and LAW_CAT_CD, no cases were designated as violation sales as there is no legal category of violation cannabis sales in New York State. So it was also dropped from the feature set. 'Lat_Lon' is not needed, as we have separate features for Latitude and Longitude. 'PD_DESC' is superfluous for the cannabis crime-only DataFrames, as the target variables describe what kind of crime each case is. 'OFNS_DESC' is also superfluous, as all cases are either labeled as 'DANGEROUS DRUGS' or 'MISCELLANEOUS PENAL LAW'. The date, time, datetime, and duration variables are all dropped as they are not numeric variables and won't be able to be processed by the classifiers. 'KY_CD' and 'PD_CD' are also dropped, as they duplicate the information stored in the target variables and will distort the fit and predict classifier functions. 'CRM_ATPT_CPTD_CD' is also dropped as 99.5% of crimes were 'completed', and this could distort the fit and predict classifier functions as well. 'X_COORD_CD' and 'Y_COORD_CD' are superfluous as latitude and longitude already exist, so they are dropped as well.
